Cho biến a và b là số nguyên

Hãy vẽ a hình tam giác đều theo hình sau, với hình trong cùng có kích cỡ các cạnh là b, các hình tiếp theo có kích cỡ tăng dần 20%



```python
import turtle as t
pen = t.Turtle()
a=int(input("Nhập số tam giác đều: "))
b=int(input("Nhập cạnh tam giác đều: "))
for i in range (1,a+1):
    for j in range (3):
        pen.forward(b)
        pen.left(120)
    if i < a:
        pen.backward(b*0.1)
    b*=1.2
t.done()
```

Cho biến a và b là số nguyên

Hãy vẽ a hình vuông với kích cỡ lớn dần theo hình sau (mỗi hình vuông có các cạnh lớn dần 10 đơn vị), hình vuông trong cùng có kích cỡ là b



```python
import turtle as t
pen = t.Turtle()
a=int(input("Nhập số hình vuông: "))
b= int(input("Nhập cạnh hình vuông: "))
for i in range (1,a+1):
    for j in range(4):
        pen.forward(b)
        pen.left(90)
    if i < a:
        pen.penup()
        pen.backward(5)
        pen.right(90)
        pen.forward(5)
        pen.left(90)
        pen.pendown()
    b+=10
t.done()
```

Cho 3 điểm toán lý hóa lưu vào 3 biến bất kỳ, tính và in ra điểm trung bình

Sau đó in ra màn hình 3 dòng chữ:

`Học sinh giỏi: {True/False}
Học sinh khá: {True/False}
Học sinh trung bình: {True/False}`

(trong đó, mỗi dòng chỉ in giá trị true hoặc false)

Giá trị true false của mỗi dòng được xét theo nguyên tắc:

Nếu điểm trung bình lớn hoặc bằng 8, đó là học sinh giỏi
Nếu điểm trung bình bé hơn 8 và lớn hơn hoặc bằng 6, đó là học sinh khá
Trường hợp còn lại là học sinh trung bình

Ví dụ, nếu điểm trung bình là 9 thì in ra 3 dòng chữ:

`Học sinh giỏi: True
Học sinh khá: False
Học sinh trung bình: False`

```python
dtoan = 9
dhoa = 9
dly = 1
dtb = (dtoan + dhoa + dly)/3
print("Học sinh giỏi: ",dtb >= 8)
print("Học sinh khá: ", 6<= dtb <8)
print("Học sinh trung bình: ",dtb<6)
```

```python
dtoan = float(input("Nhập điểm toán: "))
dhoa = float(input("Nhập điểm hóa: "))
dly = float(input("Nhập điểm lý: "))
dtb = (dtoan + dhoa + dly)/3
a = "Học sinh giỏi: "
b = "Học sinh khá: "
c = "Học sinh trung bình: "
print(a + str(dtb >= 8))
print(b + str(dtb >= 6 and dtb <8))
print(c+ str(dtb<6))
```
